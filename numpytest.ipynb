```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 0, 0, 0.1, 0, 0.1, 0])
y = np.array([0, 1, 0.5, 1, 0.5, 0, 0.5])

plt.plot(x, y)
plt.title("Sonuçlar")

plt.xlim(-0.1, 0.2)
plt.grid()
plt.show()
```
